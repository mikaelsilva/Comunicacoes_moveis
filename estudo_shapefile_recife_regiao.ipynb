```python
!pip install geopandas
import geopandas as gpd
```

```python
street_map_geojson = gpd.read_file("http://dados.recife.pe.gov.br/dataset/c1f100f0-f56f-4dd4-9dcc-1aa4da28798a/resource/e43bee60-9448-4d3d-92ff-2378bc3b5b00/download/bairros.geojson")
#street_map_shp = gpd.read_file("/content/pasta_teste/26MUE250GC_SIR.shp")
```

```python
lista_bairros = ['SAO JOSE','ILHA JOANA BEZERRA','CABANGA','SANTO ANTONIO','DERBY','ILHA DO RETIRO','ILHA DO LEITE','BOA VISTA','COELHOS','SANTO','SOLEDADE','PAISSANDU','MADALENA']
street_map = street_map_geojson[street_map_geojson["bairro_nome_ca"].isin(lista_bairros)]
street_map
```

```python
fix,ax = plt.subplots(figsize=(15,15))
street_map.plot(ax=ax, column="bairro_nome_ca")
```

```python
from shapely.geometry import Point, Polygon
```

```python
geometry = [Point(xy) for xy in zip(df["lon"],df["lat"])]
```

```python
geometry[:3]
```

```python
geo_df = gpd.GeoDataFrame(df,geometry=geometry)
```

```python
geo_df.head(10)
```

```python
fix,ax = plt.subplots(figsize = (15,15))
street_map.plot(ax=ax,alpha = 0.4, column="bairro_nome_ca")
geo_df[geo_df["pontoId"] > 0].plot(ax = ax, markersize=20, color= "red", marker = "o", label = "PontoID")
plt.legend(prop={'size':15})
```
